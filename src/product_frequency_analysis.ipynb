```python
from database_handler import just_order_items, get_item_names
import matplotlib.pyplot as plt
import collections
order_items =  just_order_items()
items_frequency = collections.Counter(order_items)
item_name_dict = get_item_names()
order_items_names = [item_name_dict[item_id] for item_id in order_items]
names_frequency = collections.Counter(order_items_names)
plt.pie(items_frequency.values(), labels = names_frequency.keys())

```
